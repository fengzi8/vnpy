```python
import talib as ta
import numpy as np
import matplotlib.pyplot as plt
import datetime
import yfinance as yf
import pandas as pd
```

```python
power = yf.Ticker("POWERGRID.NS")
df = power.history(start = '2020-01-01',end = '2020-9-4')
```

```python
df['MA'] = ta.SMA(df['Close'],20)
df[['Close','MA']].plot(figsize=(12,12))
plt.show()
```

```python
df['avg'] = ta.ADX(df['High'],df['Low'],df['Close'])
df[['avg']].plot(figsize=(12,10))
```

```python
df['up_band'],df['mid_band'],df['low_band'] = ta.BBANDS(df['Close'],timeperiod= 20)
df[['up_band','mid_band','Close','low_band'] ].plot(figsize = (12,10))
plt.show()
```
